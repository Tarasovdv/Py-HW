```python
import math
from sympy import*
from sympy.plotting import plot as plt
init_printing()

x = Symbol("x")
fx = -5*x**2 + 10*x - 150

fx
```

**Построение графика**

```python
p = plt(fx, (x,-7.5,7.5))
```

**Дифференциал**

```python
diff(fx)
```

**Корни**

```python
solve(fx)
```

**Вершина**

```python
solve(diff(fx))
```

**Функция возрастает**

```python
solve(diff(fx)>0)
```

**Функция убывает**

```python
solve(diff(fx)<0)
```

**Промежуток на котором f > 0**

```python
solve((fx>0))
```

**Промежуток на котором f < 0**

```python
solve((fx<0))
```
